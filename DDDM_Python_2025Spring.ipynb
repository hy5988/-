poisson distribution

```python
from scipy.stats import poisson

# Parameters
lambda_ = 5  # average rate (mean)

# Generate data
x = np.arange(0, 15)
y = poisson.pmf(x, lambda_)

# Plotting the PMF
plt.figure(figsize=(10, 6))
plt.bar(x, y, label='Poisson Distribution', color='green')
plt.title('Poisson Distribution (λ=5)')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.grid()
plt.legend()
plt.show()
```

Exponential Distribution

```python
from scipy.stats import expon
# Parameters
lambda_ = 5  # rate parameter (λ)

# Generate data
x = np.linspace(0, 5, 1000)
y = expon.pdf(x, scale=1/lambda_)

# Plotting the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Exponential Distribution', color='orange')
plt.fill_between(x, y, alpha=0.3)
plt.title('Exponential Distribution (λ=1)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid()
plt.legend()
plt.show()
```

Binomial Distribution

```python
from scipy.stats import binom

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success

# Generate data
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

# Plotting the PMF
plt.figure(figsize=(10, 6))
plt.bar(x, y, label='Binomial Distribution', color='orange')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid()
plt.legend()
plt.show()
```

normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 0      # mean
sigma = 1   # standard deviation

# Generate data
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Plotting the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution', color='blue')
plt.title('Normal Distribution (μ=0, σ=1)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid()
plt.legend()
plt.show()
```

sampling

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
```

```python
education_districtwise = pd.read_csv('/content/drive/MyDrive/CDT502/education_districtwise.csv')
education_districtwise = education_districtwise.dropna()
education_districtwise.shape
education_districtwise.head()
```

```python
sampled_data = education_districtwise.sample(n=50, replace=True, random_state=31208)
sampled_data.shape
sampled_data
mean_estimate = sampled_data['OVERALL_LI'].mean()
mean_estimate
```

central limit theorem

```python
estimate_list = []
for i in range(100):
  mean_sample = education_districtwise['OVERALL_LI'].sample(n=50, replace=True).mean()
  estimate_list.append(mean_sample)
```

```python
estimate_df = pd.DataFrame(data={'estimate':estimate_list})
estimate_df.shape

mean_sample_means = estimate_df['estimate'].mean()
mean_sample_means
```

```python
population_mean = education_districtwise['OVERALL_LI'].mean()
population_mean
```

```python
plt.hist(estimate_df['estimate'], bins=25, density=True, alpha=0.4, label = "histogram of sample means of 10000 random samples")
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100) # generate a grid of 100 values from xmin to xmax.
p = stats.norm.pdf(x, mean_sample_means, stats.tstd(estimate_df['estimate']))
plt.plot(x, p,'k', linewidth=2, label = 'normal curve from central limit theorem')
plt.axvline(x=population_mean, color='g', linestyle = 'solid', label = 'population mean')
plt.axvline(x=mean_estimate, color='r', linestyle = '--', label = 'sample mean of the first random sample')
plt.axvline(x=mean_sample_means, color='b', linestyle = ':', label = 'mean of sample means of 10000 random samples')
plt.title("Sampling distribution of sample mean")
plt.xlabel('sample mean')
plt.ylabel('density')
plt.legend(bbox_to_anchor=(1.04,1))
plt.show()
```

hypothesis testing

```python
education_districtwise = pd.read_csv("/content/drive/MyDrive/CDT502/education_districtwise.csv")
education_districtwise = education_districtwise.dropna()
education_districtwise.head(10)
```

one-sample t-Test

```python
state21 = education_districtwise[education_districtwise['STATNAME']=='STATE21']
state21.shape
```

```python
sample_state21 = state21.sample(n=20, replace = True, random_state = 13490)

pop_meam = 75

# perform one-sample t-Test
stats.ttest_1samp(a=sample_state21['OVERALL_LI'], popmean = pop_meam)
```

two-sample t-Test

```python
state21 = education_districtwise[education_districtwise['STATNAME']=='STATE21']
state28 = education_districtwise[education_districtwise['STATNAME']=='STATE28']

sampled_state21 = state21.sample(n=20, replace = True, random_state=13490)
sampled_state28 = state28.sample(n=20, replace = True, random_state=39103)

tstat, pvalue = stats.ttest_ind(a = sampled_state21['OVERALL_LI'] , b=sampled_state28['OVERALL_LI'] , equal_var=False)
print(tstat)
print(pvalue)
```
